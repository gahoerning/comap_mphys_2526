```python
import os
from astroquery.ipac.irsa import Irsa
from astropy.coordinates import SkyCoord
import astropy.units as u
from tqdm import trange

overwrite = False

coll = "wise_allsky"
download_path = f"/scratch/nas_comap2/hoerning/IRSA_output_jg/{coll}"

coord = SkyCoord(17.0, 0.8, frame='galactic', unit="deg")
print(f"Coordinates: ({coord.icrs})")

tbl = Irsa.query_sia(
    pos=(coord, 2*u.deg),
    collection=coll
)

print(tbl["access_url"])

# STEP 2: Then download

try:
    os.makedirs(download_path)
except FileExistsError:
    pass

import requests

f = open(f"{download_path}/download_list.txt", "a")
f.close()

print("Downloading")
for i in trange(len(tbl["access_url"])):
    url=tbl["access_url"][i]
    r = requests.get(url)
    name = url.split("/")[-1]
    path = f"{download_path}/{name}"

    # skip if exists and overwrite is False
    if path.isfile(path) and not overwrite:
        continue

    # save only .fits files
    if path.endswith(".fits"):
        n_bytes = open(path, "wb").write(r.content)
        print(f"{i}: {url} - {n_bytes} bytes")
        print(f"     - Saved to {download_path}/{name}")
        open(f"{download_path}/download_list.txt", "a").write(f"{download_path}/{name}")
```

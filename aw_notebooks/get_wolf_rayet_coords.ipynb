```python
# --- IMPORT STATEMENTS ---

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# --- GLOBAL VARIABLES ---

url = 'http://pacrowther.staff.shef.ac.uk/WRcat/index.php'
l_bounds = (16, 32) # (min, max) degrees
b_bounds = (-1.1, 1.1) # (min, max) degrees

# --- MAIN CODE ---

response = requests.get(url)

# extract table from html
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table')

# headers = []
# for th in table.find_all('th'):
#     headers.append(th.get_text(strip=True))

rows = []
for tr in table.find_all('tr')[1:]:  # skip first row (header)
    cells = tr.find_all('td')
    row = [cell.get_text(strip=True) for cell in cells]
    rows.append(row)

data = np.array(rows)

ra = np.char.replace(data[:, 7], ' ', ':')
dec = np.char.replace(data[:, 8], ' ', ':')
coords_arr = np.column_stack((ra, dec))
print(f'No. of coordinates: {len(coords_arr)}')

# # remove data points outside our map range - IN GALACTIC COORDS, NOT RA DEC
# valid_coords = []
# for coord in coords_arr:
#     l = coord[0]
#     b = coord[1]
#     if l > l_bounds[0] and l < l_bounds[1] and b > b_bounds[0] and b < b_bounds[1]: # if in our map range
#         valid_coords.append(coord)
# print(data)
# valid_coords = np.array(valid_coords)
# print(f'No. of coordinates in range {l_bounds[0]} < l < {l_bounds[1]}, {b_bounds[0]} < b < {b_bounds[1]}: {len(valid_coords)}')
# print(valid_coords)

diameter = 5/60 # 5 arcmin
output_file = 'wolf_rayet_coords.reg'
with open(output_file, 'w') as file:
    for row in coords_arr:
        file.write('fk5;circle ' + ' '.join(row) + f' {diameter}' + '\n')  # Join each row back to a string and write to the file
```

```python
def read_region(region, map):
    """
    Reads fits data within a region and outputs sum of fluxes within region
    and number of pixels in region. As far as I can tell, this is exactly
    how ds9 does it, if not then it seems to give the exact result anyways.
    """
    REGION_RADIUS = region[2]
    included_pixels = np.array([])
    # print(int(region[0] - region[2]))
    x_lbound = int(np.round(region[0]-1 - REGION_RADIUS))
    x_ubound = int(np.round(region[0] + REGION_RADIUS))
    y_lbound = int(np.round(region[1]-1 - REGION_RADIUS))
    y_ubound = int(np.round(region[1] + REGION_RADIUS))
    for x_pixel in range(x_lbound,x_ubound):
        for y_pixel in range(y_lbound,y_ubound):
            pixel_distance = np.sqrt(
                                     (region[0]-1 - x_pixel)**2 +
                                     (region[1]-1 - y_pixel)**2
                                     )
            if pixel_distance <= REGION_RADIUS:
                included_pixels = np.append(included_pixels,
                                            map[y_pixel, x_pixel])
    
    region_info = [np.sum(included_pixels), len(included_pixels)]


    return region_info
```

```python

# MAIN CODE
with open(COORD_FILE_NAME, 'r') as file:
    lines = file.readlines()
reg_data = [line.strip().split() for line in lines]
reg_data = np.delete(reg_data, 0, axis=1).astype('float') # remove 'galactic;circle' column - just l, b, radius

map_fname = MAPS_FNAMES[0] # TEMP - loop through all files later

for region in reg_data:
    print(region)
    # Open the FITS file and extract the image data (map)
    with fits.open(f'{MAPS_DIR}/{map_fname}') as hdul:
        map_data = hdul[0].data  # This assumes the image data is in the primary HDU.
    
    region_info = read_region(region, map_data)
    print(region_info)
```

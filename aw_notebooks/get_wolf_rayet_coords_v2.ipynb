```python
# --- IMPORT STATEMENTS ---

import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
```

```python
# --- GLOBAL VARIABLES ---

URL = 'http://pacrowther.staff.shef.ac.uk/WRcat/index.php'
SAVEDIR = "/home/aimee/mphys/data"
```

```python
# --- MAIN CODE ---

# Extract table from html
response = requests.get(URL)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table')

# Extract data from table
rows = []
for tr in table.find_all('tr')[1:]:  # skip first row (header)
    cells = tr.find_all('td')
    row = [cell.get_text(strip=True) for cell in cells]
    rows.append(row)
data = np.array(rows)

# Validate data, convert to float
l_arr = [float(i) for i in data[:, 9]] # degrees
b_arr = [float(i) for i in data[:, 10]] # degrees
distance_arr_raw = data[:, 29]
distance_arr = []
for i in distance_arr_raw:
    try: 
        distance_arr.append(float(i))
    except ValueError:
        distance_arr.append('nan')
# print(f'No. of coordinates: {len(l_arr)}')

# Write to csv
output_file = f'{SAVEDIR}/wolf_rayet_coords.csv'
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['l / deg', 'b / deg', 'distance / kpc']) # header
    writer.writerows(zip(l_arr, b_arr, distance_arr))
print(f'File saved to {output_file}')
```

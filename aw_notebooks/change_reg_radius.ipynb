```python
file_name = 'cornish_point_source_data.reg'

with open(file_name, 'r') as file:
    lines = file.readlines()
data = [line.strip().split() for line in lines]

diameter = 5/60 # 5 arcmin
for row in data: # replace diameters
    row[3] = f'{diameter}'

print(data)

output_name = 'cornish_point_source_data_modified.reg'
with open(output_name, 'w') as file:
    for row in data:
        file.write(' '.join(row) + '\n')  # Join each row back to a string and write to the file
```
